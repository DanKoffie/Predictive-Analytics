```python
#Load liabraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error 
```

```python
#Import data
train = pd.read_csv('C:/Users/DAN KOFFIE/Desktop/Assignment/train.csv')
test = pd.read_csv('C:/Users/DAN KOFFIE/Desktop/Assignment/test.csv')
```

```python
#View the first few rows of the dataframe
train.head(10)
```

```python
#Check data shape

train.shape
```

```python
#Check data columns 

train.columns
```

```python
#Feature selection by removing irrelevant columns

traindata=train.drop(['Name','Ticket','Cabin','PassengerId', 'Embarked'], axis=1)
traindata.head(10)
```

```python
traindata['Sex'].unique()
```

```python
#Checking for missing values

traindata.isnull().sum()
```

```python
#Convert 'Sex' column into numerical values 

traindata['Sex'] = traindata['Sex'].map({'male': 1, 'female': 0})
traindata.head()
```

```python
#Calculate mean age of each sex

mean_age_male=traindata[traindata['Sex']==1]['Age'].mean()
mean_age_female=traindata[traindata['Sex']==0]['Age'].mean()
```

```python
#Fill all  null values in 'Age' column with mean age

traindata.loc[(traindata.Age.isnull()) & (traindata['Sex']==0),'Age']=mean_age_female
traindata.loc[(traindata.Age.isnull()) & (traindata['Sex']==1),'Age']=mean_age_male
```

```python
#Check to confirm if 'Age' column is filled successfully 

traindata.isnull().sum()
```

```python
#Describing the data for summary statistics

traindata.describe()
```

```python
#Re-Scaling to standardize 'Age' and 'Fare' in range

traindata.Age = (traindata.Age-min(traindata.Age))/(max(traindata.Age)-min(traindata.Age))
traindata.Fare = (traindata.Fare-min(traindata.Fare))/(max(traindata.Fare)-min(traindata.Fare))

traindata.describe()
```

```python
#CREATING THE PREDICTION MODEL

#Getting data splitted for training and testing 

X_train, X_test, y_train, y_test = train_test_split(
    traindata.drop(['Survived'], axis=1),
    traindata.Survived,
    test_size= 0.2,
    random_state=0,
    stratify=traindata.Survived)
```

```python
#Performing logistics regression 

lrmod = LogisticRegression()
lrmod.fit(X_train, y_train)
y_predict = lrmod.predict(X_test)
accuracy_score(y_test, y_predict)
```
